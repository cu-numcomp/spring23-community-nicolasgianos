```python
from statsmodels.tsa.api import ExponentialSmoothing, Holt
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("total_cases.csv")
df = df[df["date"] >= '2020-01-20']
usableData = pd.Series(df['United States'].tolist(), pd.date_range(start = '2020-01-20', end = '2023-03-29', freq = 'D'))

fig, ax = plt.subplots(figsize = (20,8))
plt.plot(usableData);
```

```python
fit = Holt(usableData, initialization_method="estimated").fit( smoothing_level=0.595197, smoothing_trend=0.250378, optimized=False)


fig, ax = plt.subplots(figsize = (20,8))
plt.plot(fit.forecast(720)) #number of days predicted after data
plt.plot(usableData);
```
